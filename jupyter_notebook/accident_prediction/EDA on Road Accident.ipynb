```python
# Kde a plot
# # univariante analysis 
plt.figure(figsize=(15, 10))
plt.suptitle('Univariate Analysis of Numerical Features of num_features Dataset', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(num_features)):
    plt.subplot(3, 3, i+1)
    sns.histplot(x=df[num_features[i]], kde=True, color='blue')
    # Add mean line
    mean_value = df[num_features[i]].mean()
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
    
    # Add median line
    median_value = df[num_features[i]].median()
    plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
    
    plt.xlabel(num_features[i])
    plt.legend()
    plt.tight_layout()
plt.show()
```

### Observation:
1. majority of times only 2 vehicle are involved in accident 
2. on a single data majority of count of accident lies between 5 to 10 
3. most of accident happend by people whose age lies between in a range of 20 to 30 year.

# Univariate Analysis on categorical Features:

```python
import matplotlib.pyplot as plt

# Assuming df and cat_features are defined earlier
fig, axs = plt.subplots(2, 4, figsize=(20,10))  # Changed subplots to 1 row and 4 columns

plt.suptitle('Univariate Analysis of Categorical Features for dataset', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
axs_flat = axs.flatten()

# Ensure that the number of subplots matches the number of categorical features
num_plots = min(len(cat_features), len(axs_flat))

# Adjust pie size
pie_size = 1  # Adjust the size of the pie relative to the axes

for i in range(num_plots):
    feature = cat_features[i]
    value_counts = df[feature].value_counts()
    
    if len(value_counts) > 0:
        ax = axs_flat[i]
        
        # Limit to top 8 categories
        if len(value_counts) > 8:
            value_counts = value_counts[:8]
        
        ax.pie(value_counts, autopct=lambda p: '{:.2f}%'.format(p), labels=value_counts.index, radius=pie_size)
        ax.set_title(feature)
    else:
        # If no data for the feature, hide the subplot
        axs_flat[i].axis('off')

plt.tight_layout()  # Ensures the plots are properly spaced
plt.show()
```

### Observation:
1. Majority or accident happend on a dry road surface then followed by wet/damp surface.
2. majority of accident happend in daylight around 70%.
3. in normal weather condition most of accident happend. 
4. Driver has infected mostly in accident than followed by passenger and pedestrian.
5. only 12% of accident are serious and where as 0.75% accident are fatal which means death happend in accident where as 87% of accident are minor.
6. most of accident happend by Male and around 41% accident casued by Woman.
7. most of accident happend by the Car and then followed by pedal cycle and Bike

# Bivariante analysis Categorical to Numerical

```python
#Bivariante analysis 
def bar_plot_cat_vs_num(num_feature):
    plt.figure(figsize=(15, 15))
    plt.suptitle(f'Bar Plot of {num_feature} W.R.T Categorical Features of dataset', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

    for i in range(0, len(cat_features)):
        plt.subplot(3, 3, i+1)
        sns.barplot(x=df[cat_features[i]], y = df[num_feature])
        for p in plt.gca().patches:
                plt.gca().annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                                   ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                                   textcoords='offset points')
        plt.xlabel(cat_features[i])
        plt.xticks(rotation = 45)
        plt.tight_layout()
    plt.show()
```
