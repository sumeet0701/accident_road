```python
import pandas as pd
import os
import requests
import zipfile

def process_csv(year, dataframe):
    if year == 2013:
        dataframe.drop("Casualty Class", axis=1, inplace=True)
        dataframe.rename(columns={"Unnamed: 11": "Casualty Class", "Unnamed: 15": "Type of Vehicle"}, inplace=True)
    elif year == 2014:
        dataframe.drop("Casualty Class", axis=1, inplace=True)
        dataframe.rename(columns={"Unnamed: 11": "Casualty Class"}, inplace=True)
    elif year == 2016:
        dataframe.drop("Expr1", axis=1, inplace=True)
    elif year == 2018:
        columns_drop = ['Vehicle Fields_Reference Number', 'Vehicle Number', 'Local Authority',
                        'Casualty Fields_Reference Number', 'Casualty Veh No']
        dataframe.drop(columns=columns_drop, inplace=True)
        dataframe.rename(columns={"Accident Fields_Reference Number": "Reference Number"}, inplace=True)
    elif year in [2009, 2010, 2011, 2012]:
        dataframe.rename(columns={"Easting": "Grid Ref: Easting", "Northing": "Grid Ref: Northing"}, inplace=True)
        #columns_drop = ['Local Authority', 'Vehicle Number']
        #dataframe.drop(columns=columns_drop, inplace=True)
    elif year == 2019:
        columns_drop = ['Local Authority', 'Vehicle Number']
        dataframe.drop(columns=columns_drop, inplace=True)

def download_and_process_zip(url, output_csv_path):
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to download the zip file.")
        return
    
    with open("temp.zip", "wb") as zip_file:
        zip_file.write(response.content)
    with zipfile.ZipFile("temp.zip", 'r') as zip_ref:
        zip_ref.extractall("temp")
    os.remove("temp.zip")

    # Find the nested folder within the extracted directory
    extracted_folders = os.listdir("temp")
    if not extracted_folders:
        print("No folders found in the extracted directory.")
        return

    folder_path = os.path.join("temp", extracted_folders[0])

    # Process CSV files after extraction
    dfs = []
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    print("Found CSV files:", csv_files)
    if not csv_files:
        print("No CSV files found in the extracted directory.")
        return
    
    for filename in csv_files:
        year = int(filename.split(".")[0])
        df = pd.read_csv(os.path.join(folder_path, filename), encoding = "latin1")
        process_csv(year, df)
        dfs.append(df)

    # Concatenate all dataframes into a single dataframe
    combined_df = pd.concat(dfs, ignore_index=True)

    # Save the combined dataframe to a single CSV file
    combined_df.to_csv(output_csv_path, index=False)

    # Clean up temporary directory
    os.rmdir("temp")

def main():
    github_zip_url = "https://github.com/sumeet0701/data-dump/raw/main/road%20accident.zip"
    output_csv_path = "output.csv"

    # Download and process the zip file from GitHub
    download_and_process_zip(github_zip_url, output_csv_path)

if __name__ == "__main__":
    main()
```
